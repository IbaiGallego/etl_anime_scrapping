```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np

import re
```

```python
mal = pd.read_csv('../data/anime_clean.csv')
imdb = pd.read_csv('../data/imdb_full.csv')
```

```python
imdb.head()
```

```python
def title_dec(string):
    pattern = r'^\d+(?:\.\d+)*\.?\s+|\s+\(\d+(?:–.+)?\)$'
    return re.sub(pattern, '', string)
```

```python
def extract_year(string):
    pattern = r'\b\d{4}\b'
    years = re.findall(pattern, string)
    if years:
        return years[0]
    return np.nan
```

```python
imdb['year']=imdb.title_english.apply(extract_year)
imdb['title_english']= imdb.title_english.apply(title_dec)
```

```python
def extract_directors(string):
    string = str(string)
    
    directors = None
    pattern = r'Director(?:s)?:\s*([^|]*)\s*\|'
    match = re.search(pattern, string)
    if match:
        directors = match.group(1).strip()
        return directors

    else:
        return np.nan
```

```python
def extract_stars(string):
    string = str(string)
    stars = None
    pattern = r'Star[s]?:\s*([^|]+)'
    match = re.search(pattern, string)
    if match:
        stars = match.group(1).strip()
        return stars
    else:
        return np.nan
```

```python
imdb['director'] = imdb.dir_stars.apply(extract_directors)
imdb['stars'] = imdb.dir_stars.apply(extract_stars)
```

```python
imdb = imdb.drop(['dir_stars','Unnamed: 0'], axis=1)
```

```python
imdb['votes_imdb'] = imdb['votes_imdb'].fillna(0)
```

```python
non_numeric_values = imdb[imdb['votes_imdb'].astype(str).str.contains(r'[^0-9\.]')]['votes_imdb']
non_numeric_values
```

```python
indeces = [6009,15494]
imdb = imdb.drop(indeces, axis=0)
```

```python
imdb['votes_imdb'] = imdb['votes_imdb'].str.replace('.', '').astype(float)
```

```python
def fiable(x):
    if x>100000:
        return True
    else:
        return False


imdb['imdb_trust'] = imdb.votes_imdb.apply(fiable)
```
